```python
import os
import zipfile
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torch.optim as optim
from collections import Counter
from tqdm import tqdm
from torchvision import transforms, models # Added models
from torch.utils.data import Dataset, DataLoader, random_split, Subset
from PIL import Image
from torch.optim.lr_scheduler import CosineAnnealingLR

from sklearn.metrics import accuracy_score, f1_score, classification_report
from torch.cuda.amp import autocast, GradScaler

import json # For loading preprocessing info
import time # For timing epochs


# --- Define necessary variables ---
DATA_PATH = '/content/drive/MyDrive/Released_Data_NNDL_2025' # Adjust if needed
TRAIN_DATA_CSV = os.path.join(DATA_PATH, 'train_data.csv')
SUPERCLASS_MAPPING_CSV = os.path.join(DATA_PATH, 'superclass_mapping.csv')
SUBCLASS_MAPPING_CSV = os.path.join(DATA_PATH, 'subclass_mapping.csv')
TRAIN_IMAGES_ZIP = os.path.join(DATA_PATH, 'train_images.zip')
TEST_IMAGES_ZIP = os.path.join(DATA_PATH, 'test_images.zip')

# Create local directories for extracted images (use consistent paths)
LOCAL_TRAIN_PARENT_DIR = '/content/train_images_extracted' # Parent directory
LOCAL_TEST_PARENT_DIR = '/content/test_images_extracted'   # Parent directory
LOCAL_TRAIN_DIR = os.path.join(LOCAL_TRAIN_PARENT_DIR, 'train_images', 'train_images')
LOCAL_TEST_DIR = os.path.join(LOCAL_TEST_PARENT_DIR, 'test_images', 'test_images')

# Function to extract zip files (ensure it handles nested folders if necessary)
def extract_zip(zip_path, extract_to):
    """Extract a zip file to the specified directory."""
    parent_dir = os.path.dirname(extract_to)
    if not os.path.exists(parent_dir):
        os.makedirs(parent_dir)
        print(f"Created directory: {parent_dir}")
        try:
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(parent_dir) # Extract to parent directory
            print(f"Extracted {zip_path} to {parent_dir}")
            # Check if the expected subdirectory exists
            if not os.path.exists(extract_to):
                 print(f"Warning: Expected image directory {extract_to} not found after extraction.")
                 # Add logic here to potentially find the correct path if extraction is nested differently
                 possible_paths = [os.path.join(parent_dir, d) for d in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, d))]
                 if len(possible_paths) == 1:
                     print(f"Attempting to use subdirectory: {possible_paths[0]}")
                     # Update LOCAL_TRAIN_DIR / LOCAL_TEST_DIR
                 else:
                     print("Could not automatically determine correct image directory.")

        except Exception as e:
            print(f"Error extracting {zip_path}: {e}")
    else:
        print(f"Parent directory {parent_dir} already exists.")
        if not os.path.exists(extract_to):
             print(f"Image directory {extract_to} seems missing. Attempting extraction again.")
             try:
                 with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                     zip_ref.extractall(parent_dir)
                 print(f"Re-extracted {zip_path} to {parent_dir}")
             except Exception as e:
                 print(f"Error re-extracting {zip_path}: {e}")
        else:
             print(f"Image directory {extract_to} found. Skipping extraction.")


# --- Mount Drive ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    # Extract image zip files
    extract_zip(TRAIN_IMAGES_ZIP, LOCAL_TRAIN_DIR)
    extract_zip(TEST_IMAGES_ZIP, LOCAL_TEST_DIR)
except ModuleNotFoundError:
    print("Not running in Colab, skipping Drive mount and assuming data is local.")


# Load CSV data
try:
    train_data = pd.read_csv(TRAIN_DATA_CSV)
    superclass_mapping = pd.read_csv(SUPERCLASS_MAPPING_CSV)
    subclass_mapping = pd.read_csv(SUBCLASS_MAPPING_CSV)
except FileNotFoundError as e:
    print(f"Error loading CSV files: {e}")
    print("Please ensure the paths are correct and the files exist.")
    exit() # Or handle gracefully

# Calculate number of *known* classes from mappings
num_superclasses_train = len(superclass_mapping)
num_subclasses_train = len(subclass_mapping)

# Define the number of output classes including 'novel'
# Superclasses: bird, dog, reptile, novel -> num_superclasses_train + 1
# Subclasses: known_subclass_1, ..., known_subclass_N, novel -> num_subclasses_train + 1
NUM_SUPERCASSES_OUTPUT = num_superclasses_train
NUM_SUBCLASSES_OUTPUT = num_subclasses_train

# --- Define Dataset Classes ---
class MultiLabelImageDataset(Dataset):
    def __init__(self, ann_df, img_dir, transform=None):
        self.ann_df = ann_df
        self.img_dir = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.ann_df)

    def __getitem__(self, idx):
        img_info = self.ann_df.iloc[idx]
        img_name = img_info['image']
        img_path = os.path.join(self.img_dir, img_name)

        try:
            image = Image.open(img_path).convert('RGB')
        except FileNotFoundError:
             print(f"Error: Image not found at {img_path}. Skipping index {idx}.")
             # Return None or raise an error
             raise FileNotFoundError(f"Image not found: {img_path}")

        super_idx = img_info['superclass_index']
        sub_idx = img_info['subclass_index']

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(super_idx, dtype=torch.long), torch.tensor(sub_idx, dtype=torch.long)


class MultiLabelImageTestDataset(Dataset):
    def __init__(self, img_dir, transform=None):
        self.img_dir = img_dir
        self.transform = transform
        # Ensure robustness against non-image files and correct sorting
        try:
            self.image_files = sorted(
                [f for f in os.listdir(img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))],
                key=lambda x: int(os.path.splitext(x)[0]) # Sort numerically by filename (e.g., '1.jpg')
            )
        except FileNotFoundError:
            print(f"Error: Test image directory not found at {img_dir}")
            self.image_files = []
        except ValueError:
             print(f"Warning: Could not sort test files numerically in {img_dir}. Using alphabetical sort.")
             self.image_files = sorted([f for f in os.listdir(img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])


    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.img_dir, img_name)
        try:
            image = Image.open(img_path).convert('RGB')
        except FileNotFoundError:
             print(f"Error: Test image not found at {img_path}.")
             # Handle error appropriately
             return torch.zeros(3, 224, 224), img_name # Placeholder image


        if self.transform:
            image = self.transform(image)

        return image, img_name # Return image and its original filename


# --- Calculate Mean/Std (or use default ImageNet stats) ---
# Using ImageNet stats as they are robust and often used with pre-trained models
mean = torch.tensor([0.485, 0.456, 0.406])
std = torch.tensor([0.229, 0.224, 0.225])
print(f"Using ImageNet mean: {mean} and std: {std}")


# --- Define Transforms ---
def get_train_transforms(mean, std):
    return transforms.Compose([
        transforms.Resize(256), # Larger resize before crop
        transforms.RandomResizedCrop(224, scale=(0.7, 1.0)), # Adjust scale if needed
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1), # Slightly stronger jitter
        transforms.RandomRotation(20), # Increased rotation
        transforms.ToTensor(),
        transforms.Normalize(mean=mean.tolist(), std=std.tolist()) # Convert tensor to list for Normalize
    ])

def get_val_transforms(mean, std):
    return transforms.Compose([
        transforms.Resize(256), # Match resize used in training augmentation reference
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=mean.tolist(), std=std.tolist()) # Convert tensor to list
    ])

train_transforms = get_train_transforms(mean, std)
val_transforms = get_val_transforms(mean, std)

# --- Create Datasets and DataLoaders ---
try:
    # Ensure LOCAL_TRAIN_DIR exists
    if not os.path.isdir(LOCAL_TRAIN_DIR):
        raise FileNotFoundError(f"Training image directory does not exist: {LOCAL_TRAIN_DIR}")

    full_dataset = MultiLabelImageDataset(
        train_data,
        LOCAL_TRAIN_DIR,
        transform=train_transforms # Apply train transforms initially
    )

    # Split into training and validation sets
    train_size = int(0.85 * len(full_dataset)) # Use 85% for training
    val_size = len(full_dataset) - train_size
    print(f"Splitting dataset: {train_size} train, {val_size} validation")
    train_dataset, val_dataset_original = random_split(
        full_dataset,
        [train_size, val_size],
        generator=torch.Generator().manual_seed(42) # Ensure reproducibility
    )

    # Wrap the validation Subset to apply validation transforms
    class ValidationDatasetWrapper(Dataset):
        def __init__(self, subset, transform):
            self.subset = subset
            self.transform = transform
        def __getitem__(self, index):
            img, super_idx, sub_idx = self.subset[index]
            # The image received from subset already has train transforms applied.
            # We need to access the *original* image from the full_dataset
            # and apply the val_transform.

            original_index = self.subset.indices[index]
            original_img_info = self.subset.dataset.ann_df.iloc[original_index]
            img_name = original_img_info['image']
            img_path = os.path.join(self.subset.dataset.img_dir, img_name)

            try:
                 image = Image.open(img_path).convert('RGB')
            except FileNotFoundError:
                 # Handle error as before
                 raise FileNotFoundError(f"Image not found during validation wrapper: {img_path}")


            super_idx = original_img_info['superclass_index']
            sub_idx = original_img_info['subclass_index']

            if self.transform:
                image = self.transform(image) # Apply validation transform here

            return image, torch.tensor(super_idx, dtype=torch.long), torch.tensor(sub_idx, dtype=torch.long)

        def __len__(self):
            return len(self.subset)

    val_dataset = ValidationDatasetWrapper(val_dataset_original, val_transforms)


     # Ensure LOCAL_TEST_DIR exists and is correct before creating dataset
    if not os.path.isdir(LOCAL_TEST_DIR):
         print(f"Warning: Test image directory does not exist: {LOCAL_TEST_DIR}. Test loader will be empty.")

    test_dataset = MultiLabelImageTestDataset(
        LOCAL_TEST_DIR,
        transform=val_transforms # Use validation transforms for testing
    )

    # Create DataLoaders
    batch_size = 64
    num_workers = 2

    train_loader = DataLoader(
        train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers,
        pin_memory=True, persistent_workers=True if num_workers > 0 else False, drop_last=True # drop_last can help stabilize training
    )
    val_loader = DataLoader(
        val_dataset, batch_size=batch_size*2, shuffle=False, num_workers=num_workers, # Larger batch size for validation
        pin_memory=True, persistent_workers=True if num_workers > 0 else False
    )
    test_loader = DataLoader(
        test_dataset, batch_size=batch_size*2, shuffle=False, num_workers=num_workers, # Larger batch size for testing
        pin_memory=True, persistent_workers=True if num_workers > 0 else False
    )

    print(f"\nDataLoaders created:")
    print(f"  Train Loader: {len(train_loader)} batches")
    print(f"  Validation Loader: {len(val_loader)} batches")
    print(f"  Test Loader: {len(test_loader)} batches ({len(test_dataset)} images)")


except FileNotFoundError as e:
    print(f"Fatal Error during Dataset/DataLoader creation: {e}")
    print("Please check image directory paths and ensure data is extracted correctly.")
    exit()
except Exception as e:
    print(f"An unexpected error occurred during DataLoaders setup: {e}")
    exit()

# # Model Architecture
# * Uses a pre-trained ResNet-50 backbone.
# * Two separate fully connected layers (classifiers) for super-class and sub-class predictions.
# * The number of output neurons matches the number of known classes + 1 (for 'novel').

class MultiLabelClassifier(nn.Module):
    def __init__(self, num_superclasses, num_subclasses, pretrained=True):
        """
        Model with a shared backbone and separate heads for super/sub-class classification.

        Args:
            num_superclasses (int): Number of super-classes including 'novel'.
            num_subclasses (int): Number of sub-classes including 'novel'.
            pretrained (bool): Whether to use a pre-trained backbone (ResNet-50).
        """
        super().__init__()
        # Load pre-trained ResNet-50
        self.backbone = models.resnet50(pretrained=pretrained)

        # Get the number of features from the layer before the final classifier
        num_ftrs = self.backbone.fc.in_features

        # Replace the final fully connected layer with an Identity layer (we'll add our own heads)
        self.backbone.fc = nn.Identity()

        # Super-class classification head
        self.superclass_classifier = nn.Linear(num_ftrs, num_superclasses)

        # Sub-class classification head
        self.subclass_classifier = nn.Linear(num_ftrs, num_subclasses)

        print(f"Model Initialized:")
        print(f"  Backbone: ResNet-50 (Pretrained: {pretrained})")
        print(f"  Input features to classifiers: {num_ftrs}")
        print(f"  Super-class output units: {num_superclasses}")
        print(f"  Sub-class output units: {num_subclasses}")


    def forward(self, x):
        """
        Forward pass through the model.

        Args:
            x (torch.Tensor): Input image batch.

        Returns:
            tuple: (superclass_logits, subclass_logits)
        """
        # Get features from the backbone
        features = self.backbone(x)

        # Pass features through the respective classifiers
        superclass_logits = self.superclass_classifier(features)
        subclass_logits = self.subclass_classifier(features)

        return superclass_logits, subclass_logits

# Training Configuration

# --- Hyperparameters ---
LEARNING_RATE = 1e-4 # Starting learning rate
WEIGHT_DECAY = 1e-4  # Weight decay for AdamW optimizer
EPOCHS = 30          # Number of training epochs (adjust based on convergence)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

# --- Threshold for Novelty Detection ---
NOVELTY_THRESHOLD_SUPER = 0.7 # Example: If max probability < threshold, classify as novel
NOVELTY_THRESHOLD_SUB = 0.5

# --- Model Initialization ---
model = MultiLabelClassifier(
    num_superclasses=NUM_SUPERCASSES_OUTPUT,
    num_subclasses=NUM_SUBCLASSES_OUTPUT,
    pretrained=True
).to(DEVICE)

# --- Loss Function ---
# Use CrossEntropyLoss for multi-class classification
criterion_super = nn.CrossEntropyLoss()
criterion_sub = nn.CrossEntropyLoss()

# --- Optimizer ---
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# --- Learning Rate Scheduler ---
# Cosine Annealing helps in smoothly decreasing the learning rate
scheduler = CosineAnnealingLR(optimizer, T_max=EPOCHS, eta_min=1e-6) # eta_min is the minimum lr

# --- Mixed Precision Training ---
# Helps speed up training and reduce memory usage on compatible GPUs
scaler = GradScaler(enabled=torch.cuda.is_available())

# Training and Validation Functions

def train_epoch(model, dataloader, criterion_super, criterion_sub, optimizer, scaler, device):
    """Trains the model for one epoch."""
    model.train()  # Set model to training mode
    total_loss = 0.0
    correct_super = 0
    correct_sub = 0
    total_samples = 0

    progress_bar = tqdm(dataloader, desc="Training", leave=False)
    for images, super_labels, sub_labels in progress_bar:
        images, super_labels, sub_labels = images.to(device), super_labels.to(device), sub_labels.to(device)

        optimizer.zero_grad() # Zero gradients before backward pass

        # Mixed precision forward pass
        with autocast(enabled=scaler.is_enabled()):
            super_logits, sub_logits = model(images)
            loss_super = criterion_super(super_logits, super_labels)
            loss_sub = criterion_sub(sub_logits, sub_labels)
            # Combine losses
            loss = loss_super + loss_sub

        # Mixed precision backward pass
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        # Calculate Metrics
        total_loss += loss.item() * images.size(0) # Accumulate loss scaled by batch size

        _, predicted_super = torch.max(super_logits.data, 1)
        _, predicted_sub = torch.max(sub_logits.data, 1)

        total_samples += super_labels.size(0)
        correct_super += (predicted_super == super_labels).sum().item()
        correct_sub += (predicted_sub == sub_labels).sum().item()

        # Update progress bar description
        progress_bar.set_postfix(loss=loss.item(), super_acc=f"{100*correct_super/total_samples:.2f}%", sub_acc=f"{100*correct_sub/total_samples:.2f}%")

    avg_loss = total_loss / total_samples
    avg_super_acc = 100 * correct_super / total_samples
    avg_sub_acc = 100 * correct_sub / total_samples

    return avg_loss, avg_super_acc, avg_sub_acc


def validate_epoch(model, dataloader, criterion_super, criterion_sub, device, novelty_threshold_super, novelty_threshold_sub, num_known_super, num_known_sub):
    """Validates the model on the validation set."""
    model.eval()  # Set model to evaluation mode
    total_loss = 0.0
    correct_super = 0
    correct_sub = 0
    total_samples = 0

    all_super_preds = []
    all_super_labels = []
    all_sub_preds = []
    all_sub_labels = []

    # Indices representing the 'novel' class
    novel_super_idx = num_known_super
    novel_sub_idx = num_known_sub

    progress_bar = tqdm(dataloader, desc="Validation", leave=False)
    with torch.no_grad(): # Disable gradient calculation for validation
        for images, super_labels, sub_labels in progress_bar:
            images, super_labels, sub_labels = images.to(device), super_labels.to(device), sub_labels.to(device)

            # Forward pass
            super_logits, sub_logits = model(images)

            loss_super = criterion_super(super_logits, super_labels)
            loss_sub = criterion_sub(sub_logits, sub_labels)
            loss = loss_super + loss_sub

            total_loss += loss.item() * images.size(0)
            total_samples += super_labels.size(0)

            # Apply Novelty Thresholding
            super_probs = F.softmax(super_logits, dim=1)
            sub_probs = F.softmax(sub_logits, dim=1)

            max_super_prob, predicted_super_raw = torch.max(super_probs, 1)
            max_sub_prob, predicted_sub_raw = torch.max(sub_probs, 1)

            # Classify as novel if max probability is below threshold
            predicted_super = torch.where(max_super_prob < novelty_threshold_super, novel_super_idx, predicted_super_raw)
            predicted_sub = torch.where(max_sub_prob < novelty_threshold_sub, novel_sub_idx, predicted_sub_raw)

            # Calculate Metrics
            correct_super += (predicted_super == super_labels).sum().item()
            correct_sub += (predicted_sub == sub_labels).sum().item()

            # Store predictions and labels for detailed report
            all_super_preds.extend(predicted_super.cpu().numpy())
            all_super_labels.extend(super_labels.cpu().numpy())
            all_sub_preds.extend(predicted_sub.cpu().numpy())
            all_sub_labels.extend(sub_labels.cpu().numpy())

            progress_bar.set_postfix(loss=loss.item())


    avg_loss = total_loss / total_samples

    avg_super_acc = 100 * correct_super / total_samples
    avg_sub_acc = 100 * correct_sub / total_samples

    return avg_loss, avg_super_acc, avg_sub_acc


# Main Training Loop

best_val_acc = 0.0 # Track best validation accuracy
best_epoch = -1
history = {'train_loss': [], 'val_loss': [], 'train_super_acc': [], 'val_super_acc': [], 'train_sub_acc': [], 'val_sub_acc': []}
model_save_path = '/content/best_model.pth' # Path to save the best model

print("\nStarting Training...")
start_time = time.time()

for epoch in range(EPOCHS):
    epoch_start_time = time.time()

    # --- Training ---
    train_loss, train_super_acc, train_sub_acc = train_epoch(
        model, train_loader, criterion_super, criterion_sub, optimizer, scaler, DEVICE
    )

    # --- Validation ---
    val_loss, val_super_acc, val_sub_acc = validate_epoch(
        model, val_loader, criterion_super, criterion_sub, DEVICE,
        NOVELTY_THRESHOLD_SUPER, NOVELTY_THRESHOLD_SUB, # Pass thresholds
        num_superclasses_train, num_subclasses_train     # Pass known class counts
    )

    # --- Learning Rate Step ---
    scheduler.step()

    # --- Logging ---
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['train_super_acc'].append(train_super_acc)
    history['val_super_acc'].append(val_super_acc)
    history['train_sub_acc'].append(train_sub_acc)
    history['val_sub_acc'].append(val_sub_acc)

    epoch_time = time.time() - epoch_start_time
    print(f"\nEpoch {epoch+1}/{EPOCHS} [{epoch_time:.2f}s] - "
          f"LR: {optimizer.param_groups[0]['lr']:.6f}")
    print(f"  Train Loss: {train_loss:.4f}, Super Acc: {train_super_acc:.2f}%, Sub Acc: {train_sub_acc:.2f}%")
    print(f"  Val Loss:   {val_loss:.4f}, Super Acc: {val_super_acc:.2f}%, Sub Acc: {val_sub_acc:.2f}%")


    # --- Save Best Model ---
    # Save based on validation superclass accuracy + validation subclass accuracy
    if val_super_acc + val_sub_acc > best_val_acc:
        best_val_acc = val_super_acc + val_sub_acc
        best_epoch = epoch + 1
        torch.save(model.state_dict(), model_save_path)
        print(f"  -> Best model saved to {model_save_path} (Val Super Acc: {best_val_acc:.2f}%)")

training_time = time.time() - start_time
print(f"\nTraining Finished in {training_time/60:.2f} minutes.")
print(f"Best Validation Superclass Accuracy: {best_val_acc:.2f}% at Epoch {best_epoch}")


# Plotting Training History

plt.figure(figsize=(12, 5))

# Loss Plot
plt.subplot(1, 2, 1)
plt.plot(range(1, EPOCHS + 1), history['train_loss'], label='Train Loss')
plt.plot(range(1, EPOCHS + 1), history['val_loss'], label='Validation Loss')
plt.title('Loss vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Accuracy Plot (Superclass)
plt.subplot(1, 2, 2)
plt.plot(range(1, EPOCHS + 1), history['train_super_acc'], label='Train Superclass Acc')
plt.plot(range(1, EPOCHS + 1), history['val_super_acc'], label='Validation Superclass Acc')
plt.plot(range(1, EPOCHS + 1), history['train_sub_acc'], label='Train Subclass Acc', linestyle='--')
plt.plot(range(1, EPOCHS + 1), history['val_sub_acc'], label='Validation Subclass Acc', linestyle='--')
plt.title('Accuracy vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig("/content/training_plots.png") # Save the plot
plt.show()


# # Testing Phase
# * Load the best model saved during training.
# * Run inference on the `test_loader`.
# * Apply the novelty detection logic using the chosen thresholds.
# * Format predictions into a `submission.csv` file.

# %%
def test_model(model, dataloader, device, novelty_threshold_super, novelty_threshold_sub, num_known_super, num_known_sub, best_model_path):
    """Generates predictions on the test set using the best model."""

    # Load the best model state
    try:
        model.load_state_dict(torch.load(best_model_path, map_location=device))
        print(f"Loaded best model weights from {best_model_path}")
    except FileNotFoundError:
        print(f"Error: Best model file not found at {best_model_path}. Cannot perform testing.")
        return None
    except Exception as e:
        print(f"Error loading model weights: {e}")
        return None


    model.eval()  # Set model to evaluation mode
    predictions = []
    image_filenames = []

    # Indices representing the 'novel' class
    novel_super_idx = num_known_super
    novel_sub_idx = num_known_sub

    progress_bar = tqdm(dataloader, desc="Testing", leave=False)
    with torch.no_grad():
        for images, filenames in progress_bar:
            images = images.to(device)

            # Forward pass
            super_logits, sub_logits = model(images)

            # Apply softmax to get probabilities
            super_probs = F.softmax(super_logits, dim=1)
            sub_probs = F.softmax(sub_logits, dim=1)

            # Get the prediction based on max probability
            max_super_prob, predicted_super_raw = torch.max(super_probs, 1)
            max_sub_prob, predicted_sub_raw = torch.max(sub_probs, 1)

            # Apply novelty threshold
            predicted_super = torch.where(max_super_prob < novelty_threshold_super, novel_super_idx, predicted_super_raw)
            predicted_sub = torch.where(max_sub_prob < novelty_threshold_sub, novel_sub_idx, predicted_sub_raw)

            # Store results
            predictions.extend(zip(predicted_super.cpu().numpy(), predicted_sub.cpu().numpy()))
            image_filenames.extend(filenames) # Store the filenames passed by the test dataloader


    # --- Create Submission DataFrame ---
    if not image_filenames:
        print("Warning: No image filenames were collected during testing. Cannot create submission file.")
        return None
    if len(image_filenames) != len(predictions):
         print(f"Warning: Mismatch between number of filenames ({len(image_filenames)}) and predictions ({len(predictions)}). Submission file might be incorrect.")


    submission_df = pd.DataFrame({
        'image': image_filenames,
        'superclass_prediction': [pred[0] for pred in predictions],
        'subclass_prediction': [pred[1] for pred in predictions]
    })

    # Ensure 'image' column is just the filename (ex. '1.jpg') not the full path

    # Sort by image ID
    try:
        submission_df['image_id'] = submission_df['image'].apply(lambda x: int(os.path.splitext(x)[0]))
        submission_df = submission_df.sort_values(by='image_id').drop(columns=['image_id'])
        print("Submission file sorted by image ID.")
    except ValueError:
        print("Warning: Could not sort submission file by numeric image ID. Using original order.")
    except Exception as e:
        print(f"Error during submission sorting: {e}. Using original order.")

    return submission_df

# --- Run Testing ---
print("\nStarting Testing Phase...")
# Make sure test_loader is not empty
if len(test_loader) > 0:
    submission_df = test_model(
        model,
        test_loader,
        DEVICE,
        NOVELTY_THRESHOLD_SUPER, # Use the same threshold as validation
        NOVELTY_THRESHOLD_SUB,
        num_superclasses_train,
        num_subclasses_train,
        model_save_path # Path to the best model saved during training
    )

    if submission_df is not None:
        submission_file_path = '/content/submission.csv'
        submission_df.to_csv(submission_file_path, index=False)
        print(f"\nSubmission file created successfully at: {submission_file_path}")
        print(submission_df.head())

        # Check distribution of predicted novel classes
        novel_super_count = (submission_df['superclass_prediction'] == num_superclasses_train).sum()
        novel_sub_count = (submission_df['subclass_prediction'] == num_subclasses_train).sum()
        print(f"Predicted 'novel' superclass count: {novel_super_count} / {len(submission_df)}")
        print(f"Predicted 'novel' subclass count: {novel_sub_count} / {len(submission_df)}")

else:
    print("Test loader is empty or test directory was not found. Skipping testing.")

print("\nProject Execution Complete.")
```

```python
# Convert Submission Indices to Class Names

import pandas as pd
import os

print("Converting submission indices to class names...")

# --- File Paths ---
SUBMISSION_FILE = '/content/submission.csv'
SUPERCLASS_MAPPING_CSV = os.path.join(DATA_PATH, 'superclass_mapping.csv') # Reuse from previous block
SUBCLASS_MAPPING_CSV = os.path.join(DATA_PATH, 'subclass_mapping.csv')     # Reuse from previous block
OUTPUT_NAMED_SUBMISSION_FILE = '/content/drive/MyDrive/NNDL_submission_NovelFix_DbAcc_Named.csv'

# --- Load Data ---
try:
    submission_df = pd.read_csv(SUBMISSION_FILE)
    superclass_map_df = pd.read_csv(SUPERCLASS_MAPPING_CSV)
    subclass_map_df = pd.read_csv(SUBCLASS_MAPPING_CSV)
    print("Loaded submission and mapping files.")
except FileNotFoundError as e:
    print(f"Error loading files: {e}")
    print("Please ensure submission.csv exists and mapping file paths are correct.")
    exit()

# --- Create Index-to-Name Mappings ---

# Superclass Mapping
# Create a dictionary {index: class_name}
superclass_idx_to_name = pd.Series(superclass_map_df['class'].values, index=superclass_map_df['index']).to_dict()

novel_super_idx = len(superclass_idx_to_name)
superclass_idx_to_name[novel_super_idx] = 'novel'
print(f"Superclass mapping created. Novel index: {novel_super_idx}")

# Subclass Mapping
# Create a dictionary {index: class_name}
subclass_idx_to_name = pd.Series(subclass_map_df['class'].values, index=subclass_map_df['index']).to_dict()
# Add the 'novel' class mapping.
novel_sub_idx = len(subclass_idx_to_name)
subclass_idx_to_name[novel_sub_idx] = 'novel'
print(f"Subclass mapping created. Novel index: {novel_sub_idx}")


# --- Apply Mappings to Submission DataFrame ---

# Use the .map() function to replace indices with names
submission_df['superclass_name'] = submission_df['superclass_prediction'].map(superclass_idx_to_name)
submission_df['subclass_name'] = submission_df['subclass_prediction'].map(subclass_idx_to_name)

# Handle any potential indices that weren't in the map (shouldn't happen with 'novel' added)
submission_df['superclass_name'].fillna('Unknown_Superclass', inplace=True)
submission_df['subclass_name'].fillna('Unknown_Subclass', inplace=True)


# --- Create Final Named Submission File ---

# Select and order columns for the final output
final_named_df = submission_df[['image', 'superclass_name', 'subclass_name']]

# Save the DataFrame with names
final_named_df.to_csv(OUTPUT_NAMED_SUBMISSION_FILE, index=False)

print(f"Successfully created named submission file: {OUTPUT_NAMED_SUBMISSION_FILE}")
print("\nFirst 5 rows of the named submission file:")
print(final_named_df.head())
```
